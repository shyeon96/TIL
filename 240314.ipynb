## App URL mapping
- 각 앱에 URL을 정의하는 것
- 프로젝트와 각 앱이 URL을 나누어 관리를 편하게 하기 위함
- 앱 내에 url 파일을 만들어 관리

### include()
- 프로젝트 내부 앱들의 URL을 참조할 수 있도록 매핑하는 함수
- URL의 일치하는 부분까지 잘라내고 남은 문자열 부부는 후속 처리를 위해 include된 URL로 전달

## Django Model
> DB의 테이블을 정의하고 데이터를 조작할 수 있는 기능들을 제공  
> 테이블 구조를 설계하는 청사진  

### model 클래스 작성
```python
class Article(models.Model):
    title = models.CharField()  
    content = models.TextField()
```
- title ,content : 테이블의 각 필드 명
- model field 클래스 : 테이블 필드의 데이터 타입
- django.db.models 모듈의 Model 이라는 부모 클래스를 상속받음
- Model은 model에 관련된 모든 코드가 이미 작성 되어있는 클래스

## Migrations
- model 클래스의 변경사항을 DB에 최종 반영하는 방법

### migration 과정
- model class (설계도 초안)
- migration 파일 (최종설계도)
- db.sqlite3 (DB)

### migration 핵심명령어
- python manage.py makemigrations
    - model class를 기반으로 최종 설계도 작성
- python manage.py migrate
    - 최종 설계도를 DB에 전달하여 반영

### Model Field
- DB 테이블의 열을 정의하며, 해당 필드에 저장되는 데이터 타입과 제약조건을 정의

#### CharField()
- 길이의 제한이 있는 문자열을 넣을때 사용 (max_length 는 필수 인자)

#### TextField()
- 글자의 수가 많을 때 사용

#### DateTimeField()
- 날짜와 시간을 넣을 때 사용
