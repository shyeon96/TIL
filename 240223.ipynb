## 비트연산
- AND와 OR 비트연산자 이해하기
    - a AND b => a & b
    - a OR b => a | b
    - a XOR b (같으면 0 다르면 1) => a ^ b
        - 어떤 값이던 임의의 수로 2번 XOR하면 원래수로 돌아온다
        - 이러한 성질을 이용해 암호화에 사용할 수 있다

    - Left, Right shift 연산자
        - left shift : 특정 수 만큼 비트를 왼쪽으로 밀어낸다 (우측에 0이 생성된다)
            - num << 2  # 2회 이동
        - right shift : 특정 수 만큼 비트를 오른쪽으로 밀어낸다 (우측 비트 제거)
            - num >> 2 
    
    - NOT 연산자
        - 모든 비트를 반전시킨다

- 파이썬에서 2진수 16진수 10진수 변환하여 출력하기
    - 2진수는 숫자 0 과 소문자 b -> 0b를 접두사로 붙여 표현한다
    - 16진수는 숫자 0과 소문자 x -> 0x를 접두사로 붙여 표현한다

- 음수 표현 방법
    - 컴퓨터는 음수를 2의 보수로 관리한다
    - 맨 앞자리 bit(MSB)는 음수 or 양수를 구분하는 비트이다 (1이면 음수 0이면 양수)


## 실수
- 파이썬은 f-string 문법을 지향한다

- 소수점 출력 방법
    - {n:.2f}: n값을 소수 2번째자리까지 반올림
- 컴퓨터는 실수를 내부에서 근사적으로 관리한다
    - 정확한 값이 아니라 근사 값으로 저장되는데 이때 생기는 작은 오차가 계산과정에서 다른 결과를 가져온다
