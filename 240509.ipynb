# State Management

## 컴포넌트 구조의 단순화
- 상태
    - 앱 구동에 필요한 기본 데이터
- 뷰
    - 상태를 선언적으로 매핑하여 시각하
- 기능
    - 뷰에서 사용자 입력에 대해 반응적으로 상태를 변경할 수 있게 정의된 동작

## 상태관리의 단순성이 무너지는 시점
- 계층 구조가 깊어질 경우 비효율적, 관리가 어려워 짐
- 서로 다른 뷰의 기능이 동일한 상태를 변경시켜야 하는 경우

> 각 컴포넌트의 공유 상태를 추출하여 전역에서 참조할 수 있는 저장소에서 관리
