# 240320

## 그래프
- 그래프는 정점들의 집합과 이들을 연결하는 간선들의 집합으로 구성된 자료구조
- 선형 자료구조나 트리 자료구조로 표현하기 어려운 N:N 관계를 가지는 원소들을 표현하기에 용이하다.

## 그래프 유형
- 무향 그래프
- 유향 그래프
- 가중치 그래프
- 사이클이 없는 방향 그래프
- 완전 그래프
    - 정점들에 대해 가능한 모든 간선들을 가진 그래프
- 부분 그래프
    - 원래 그래프에서 일부의 정점이나 간선을 제외한 그래프

### 인접정점
- 두 개의 점점에 간선이 존재하면 서로 인접해있다고 한다

## 그래프 표현
- 인접행렬
    - 2차원 배열을 이용해서 간선의 정보를 저장
    - 배열의 배열
- 인접 리스트
    - 각 정점마다 해당 정점으로 나가는 간선의 정보를 저장
- 간선의 배열
    - 간선을 배열에 연속적으로 저장

## Union-Find (서로소 집합)
- 서로소 또는 상호배타 집합들은 서로 중복 포함된 원소가 없는 집합들이다.
- 상호배타 집합을 표현하는 방법
    - 연결 리스트
    - 트리
- 상호배타 집합 연산
    - Make-Set(x)
    - Find-Set(x)
    - Union(x, y)
